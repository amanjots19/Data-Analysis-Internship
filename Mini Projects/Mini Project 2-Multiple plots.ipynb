```python
import pandas as pd
unrate=pd.read_csv(r"C:\Users\Gurkirat Singh\Documents\Unrate\unrate.csv")
unrate.head()
```

```python
unrate['DATE']= pd.to_datetime(unrate['DATE'])
```

```python
#Five months data
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12,12))
for i in range(0,5):
    ax=fig.add_subplot(5,1,i+1)
    start_index=i*12
    end_index=(i+1)*12
    subset=unrate[start_index:end_index]
    plt.xticks(rotation=90)
    ax.plot(subset["DATE"],subset["VALUE"])
plt.show()
```

```python
#extracting months from DATES.
unrate["MONTHS"]=unrate["DATE"].dt.month
```

```python
#Combining all plots in one.
fig = plt.figure(figsize=(6,6))
plt.plot(unrate[0:12]['MONTHS'], unrate[0:12]['VALUE'],c='red',label='1948')
plt.plot(unrate[12:24]['MONTHS'], unrate[12:24]['VALUE'],c='blue',label='1949')
plt.plot(unrate[24:36]['MONTHS'], unrate[24:36]['VALUE'],c='green',label='1950')
plt.plot(unrate[36:48]['MONTHS'], unrate[36:48]['VALUE'],c='orange',label='1951')
plt.plot(unrate[48:60]['MONTHS'], unrate[48:60]['VALUE'],c='black',label='1952')
plt.title('Unemployment Trend, 1948')
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.legend(loc='upper left')
plt.show()
```
