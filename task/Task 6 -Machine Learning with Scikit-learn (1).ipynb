```python
import pandas as pd

df=pd.read_csv(r"C:\Users\Gurkirat Singh\Documents\Diamonds Data set\diamonds.csv",index_col=0)
df.head()
```

```python
#I would like to start us using linear regression, so it's also fairly ideal that our string classifications are linear, meaning they have a meaningful order
df["cut"].unique()
```

```python
cut_class_dict={"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
```

```python
df["clarity"].unique()
```

```python
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J": 1,"I": 2,"H": 3,"G": 4,"F": 5,"E": 6,"D": 7}
```

```python
df['cut']=df['cut'].map(cut_class_dict)
df["clarity"]=df["clarity"].map(clarity_dict)
df["color"]=df["color"].map(color_dict)
df.head()
```

```python
import sklearn 
from sklearn.linear_model import SGDRegressor

df=sklearn.utils.shuffle(df)

X=df.drop("price", axis=1).values
y=df["price"].values
```

```python
test_size=200
X_train=X[:-test_size]
y_train=y[:-test_size]

X_test=X[-test_size:]
y_test=y[-test_size:]
```

```python
clf=SGDRegressor(max_iter=1000)
clf.fit(X_train,y_train)

print(clf.score(X_test,y_test))
```

```python
#Hence we're getting our result in neg , that is not possible in actual so let's change our regression model
import sklearn 
from sklearn import svm
clf=svm.SVR(kernel="rbf")
clf.fit(X_train,y_train)

print(clf.score(X_test,y_test))
```

```python
#Let's print our regression model prediction.
for X,y in zip(X_test,y_test):
    print(f"model predicts {clf.predict([X])[0]}, Actual {y}")
```
